```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
```

```python

url = requests.get('https://prodesigntools.com/tag/ddl/page/2')
page = BeautifulSoup(url.content, 'html.parser')
```

```python
h2 = page.find_all('h2', class_='entry-title')
urls = []
for i in h2:
    a = i.find('a', href=True)
    urls.append(a['href'])

urls
```

```python
tools_url = urls[1]
tools_get = requests.get(tools_url)
tools_page = BeautifulSoup(tools_get.content, 'html.parser')
```

```python
links = tools_page.find('table', class_='m-font centered')
```

```python
th = links.find_all('th')


for i in th[:3]:
    print(i.text)
```

```python
download_links = links.find_all('a')
for i in download_links:
    print(i['href'])
```

```python
table = tools_page.find('div', class_='entry-content').find_all('table')[1]

tbody = table.find('tbody').find_all('tr')

links = []

for i in tbody:
    try:
        lnk = i.a['href']
        if lnk is not None:
            links.append(lnk)
            
        
    except:
        links.append('Link not Found')

len(links)
```

```python
names = []


for i in tbody:
    try:
        b = i.find('td').text
        
        if b is not None:
            names.append(b)
    except:
        continue
len(names)
```

```python
size = []


for i in tbody:
    try:
        b = i.find('td', {'style':'text-align:right;'})
        for j in b:
            if b is not None:
                size.append(b.text)
    except:
        continue
    

len(size)
```

```python
adobe_tools = ['Product Names','Product Size','Product Download Link']


Tools = pd.DataFrame(list(zip(names,size,links)), columns=['Produc Names','Product Size','Product Download Links'])

Tools.index = np.arange(1, len(Tools) + 1)
# df.index = arange( 1, len(df) + 1)
Tools
```

```python
Tools.to_csv('Adobe CC 2019 Products Download Links.csv', index_label='S.No')
```
